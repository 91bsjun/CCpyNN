```python
import sys
import numpy as np
import pandas as pd

sys.path.append("D:\\GitHub\\CCpyNN")
from CCpyNN.StructureEncoder import atomic_info, numerical_onehot_encoder

from pymatgen.core.periodic_table import Element

atomic_info = atomic_info()

print("Original data")
sites_data_ori = pd.read_csv("Data\\sites_info_data.csv")
print(sites_data_ori.head())

```

## Numeric data (num of Li, volume) 을 One hot encoding

```python
numeric_data = sites_data_ori.copy()
numeric_data = numeric_data[['num_of_li', 'volume']]

tmp = []
for val in numeric_data['num_of_li']:
    tmp.append(numerical_onehot_encoder(20, 32, 45, val))
numeric_data['num_of_li'] = tmp

tmp = []
for val in numeric_data['volume']:
    tmp.append(numerical_onehot_encoder(numeric_data['volume'].min(), numeric_data['volume'].max(), 45, val))
numeric_data['volume'] = tmp

print(numeric_data.head())
```

## 각 sites element symobl 을 atomic information matrix 로 변환

```python
print("\nextract sites info only")
sites_data = sites_data_ori.copy()
sites_data = sites_data[["16e", "4b_1", "4b_2", "4b_3", "4b_4", "4a_1", "4a_2", "4a_3", "4a_4", "4c_1", "4c_2", "4c_3", "4c_4"]]
print(sites_data.head())

print("\nAtomic to atomic info")
for key in sites_data.keys():
    tmp = []
    for val in sites_data[key]:
        val = Element(val).Z
        tmp.append(atomic_info[val-1])
    sites_data[key] = tmp
print(sites_data.head())
```

## 만든 데이터 합치기

```python
y = sites_data_ori.copy()
y = y[['conductivity']]
concat_data = pd.concat([sites_data, numeric_data, y], axis=1)
print(concat_data.head())
```

```python
data = concat_data.copy()
data = data.sample(n=len(data))

y = data.pop("conductivity")
X = data.copy()
tmp = []
for key in X.keys():
    tmp.append(X[key])

print(X.shape)
print(X[0][0].shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D

model = Sequential()
model.add(Conv2D(64, (3, 3), input_shape=X.shape[1:]))  # shape [1:] : -1 부분 제외
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
```
